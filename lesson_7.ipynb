```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python,], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'Num_Respondents', 'LanguageHaveWorkedWith': 'Num_Knows_Python'}, inplace=True)
python_df
```

```python
python_df['Pct_Knows_Python'] = (python_df['Num_Knows_Python'] / python_df['Num_Respondents']) * 100
python_df
```

```python
python_df.sort_values(by='Pct_Knows_Python', ascending=False, inplace=True)
python_df.loc['Japan']
```
